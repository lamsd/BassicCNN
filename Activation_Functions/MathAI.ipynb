```python
confx = (10, 5)
def displays(x, y, title):
    plt.style.use('seaborn')
    plt.figure(figsize=confx)
    plt.xlabel("Weighted sum of inputs")
    plt.ylabel("Output")
    plt.plot(x, y)
    plt.title('{}'.format(title))
    plt.legend()
    plt.show()
```

```python
confx = (10, 5)
def displays2plot(x1, y1, x2, y2, label1, label2, title):
    plt.style.use('seaborn')
    plt.figure(figsize=confx)
    plt.xlabel("Weighted sum of inputs")
    plt.ylabel("Output")
    plt.plot(x1, y1, 'r', label=label1)
    plt.plot(x2, y2, 'b', label=label2)
    plt.title('{}'.format(title))
    plt.legend()
    plt.show()
```

```python
confx = (10, 5)
def axis_displays2plot(x1, y1, x2, y2, label1, label2, title):
    plt.style.use('seaborn')
    plt.figure(figsize=confx)
    plt.xlabel("Weighted sum of inputs")
    plt.ylabel("Output")
    plt.plot(x1, y1, 'r', label=label1)
    plt.plot(x2, y2, 'b', label=label2)
    plt.title('{}'.format(title))
    plt.legend()
    plt.ylim(-5, 5)
    plt.show()
```

## 1. Binary Step Function

\begin{equation*}
  f(x) = \left\{
    \begin{aligned}
      & 1, x \geq  0 \\
      & 0, x < 0
    \end{aligned}
  \right.
\end{equation*}

### Key features:
- Known the threshold activation function.
- Ouput is a binary value, either 0 or 1.
- Using for the tw linear components. Piecewise linear funtion, binary step function is a non-linear function.

### Usage:
- Do not usually use this function in modern neural network models.
- This will explain theoretical concept such as "firing a neron", "innner working of a perceptron .

```python
def binary_step(x):
    if x < 0:
        return -1
    else:
        return 1
t = np.linspace(-100, 100)
y = np.asarray([binary_step(i) for i in t])
title = 'Binary Sigmoid Activation Function'
displays(t, y, title)
```

## 3.Sigmoid

$$sigmoid(z) = \frac {1}{1 + e^{-z}}$$

### Key features:
- This is also called the logistic fucntion used in logistic regression models.
- The sigmoid function has an  s-shape graph.
- clearly, this is a non-linear function.
- The sigmoid fucntion converts its input  into a probability value between 0 and 1.
- It converts large negative valus towards 0 and large positive values towards 1.
- It returns 0.5 for the input 0. The value 0.5 is known as the threshold value which can decide that a given input belongs to what type of two classes.

### Usage:
- In the early days, the sigmoids function was used as an activation function for the hidden layers in MLPs, CNNs and RNNs.
- However, the sigmoid function is still used in RNNs.
- Currently, we do not usually use the sigmoid function for the hidden layers in MLPs and CNNs. Instead, we use ReLU and Leaky ReLU there.
- The sigmoid function must be used in the output layer when  we build a binary classifier in which  the output  is interpreted as a class label depending on the probability value of input returned by the function.
- The sigmoid fucntion  is used when we build a multilabel  classification  model in which  each mutually  inclusive class has two outcomes. Do not confuse this with  a multiclass classification model.
![image](Image/1.png)

### Drawbacks:
-  The sigmoid function has the vanshing  gradient problem. This is also known as saturation of the gradients.
- The sigmoid function has low convergence.
- Its outputs are not zero-centered. Therefore, it makes the optimization process harder.
- This function is computationally expensive as an $e^{z}$ term is include.

```python
def sigmoid_function(x):
    z = (1/(1+np.exp(-x)))
    return z

def dev_sigmoid_function(x):
    z = sigmoid_function(x)*(1-sigmoid_function(x))
    return z

t = np.linspace(-20, 20)
y = sigmoid_function(t)
y2 = dev_sigmoid_function(t)
title = 'Sigmoid Activation Function'
displays2plot(t, y, t, y2, 'sigmoid(x)', '(sigmoid(x))\'', title)
```

## 4.Tanh

$$tanh(x) = 2sigmoid(2x) -1 $$ 

$$tanh(x) = \frac{2}{(1+e^{-2x})}-1$$

$$(tanh(x))' = 1- tanh^{2}(x) = 1 - (1-2sigmoid(2x))^{2} = 4*sigmoid(2x)(1-sigmoid(2x)) = \frac{4}{(1+e^{-2x})}(1- \frac{1}{(1+e^{-2x})})  $$

### Key features.
- The output of the tanh (tangent hyperbolic) function  always ranges between -1 and +1.
- Like the sigmoid function, it has an s-shaped graph. This is also a non-linear fucntion.
- One advantage of using the tanh function over the sigmoid function is that  the tanh function is zero centered. This makes the optimization process much easier.
- The tanh function has a steeper gradient than the sigmoid fucntion has.

### Usage:
- Until recently, the tanh function was used as an activation function for the hidden layers in MLPs, CNNs and RNNs.
- Howwever, the tanh function is still  used in RNNs.
- Currently, we do not usually use the tanh fucntion for the hidden layer in MLPs and CNNs. Instead, we use ReLU or Leaky ReLU there.
- We never use the tanh function in the output layer.

### Drawbacks:
- The tanh fucntion has the vanishing gradient problem.
- This funtion is computationally expensive as an $e^{z}$ term is included.

```python
def tanh_function(x):
    z  = (2/(1+np.exp(-2*x))-1)
    return z
def dev_tanh_function(x):
    z = 1- tanh_function(x)**2
    return z

t = np.linspace(-20, 20)
y = tanh_function(t)
y2 = dev_tanh_function(t)
title = 'Tanh Activation Function'
displays2plot(t, y, t, y2, 'tanh(x)', '(tanh(x))\'', title)
```

## 5.ReLU

$$f(x) = max(0, x)$$

\begin{equation*}
  f(x) = \left\{
    \begin{aligned}
      & x, x \geq  0 \\
      & 0, x < 0
    \end{aligned}
  \right.
\end{equation*}

\begin{equation*}
  f'(x) = \left\{
    \begin{aligned}
      & 1, x \geq  0 \\
      & 0, x < 0
    \end{aligned}
  \right.
\end{equation*}

### Key features:
- The ReLU(Rectified Linear Unit) activation function is a great alternative to both sigmoid and tanh activation functions.
- Inventing ReLU is one of the most important breakthroughs made in deep learning.
- This function does not have the vansihing gradient problem. 
- This function is computationally inexpensive. It is considered that the converrgence of ReLU is 6 times faster than sigmoid and tanh functions.
- This the input  values is 0 or greater than 0, the ReLU function outputs the input as it is. If the input is less than 0, the ReLU function outputs the value 0.
- The ReLU function is made up of two linear components. Because of that, the ReLU function is a piecewise linear function. In fact, the ReLU function is a non-linear function.
- The output of the ReLU function can range from 0 to positive infinity
- The convergence is faster than sigmoid and tanh functions. This is because the ReLU function has a fixed derivate (slope) for one linear component and a zero derivative for the other linear component. Therefore, the learning process is much faster with the ReLU function.
- Calculations can be performed much faster with ReLU because no exponential terms are included in the function.

### Usage:
- The ReLU function  is the default in modern MLP and CNNs models.
- In RNNs moldel, we do not usually use it. Instead, usning sigmoid and tanh function.
- Nerver, using  the ReLU function in the output layer.

### Drawbacks:
- a dying ReLU problem.
- The value of the positive side can go very high. That may lead to a computational issue during the training

```python
def relu_function(x):
    if x < 0:
        return 0
    else:
        return x
def dev_relu_function(x):
    if x < 0:
        return 0
    else:
        return 1
    
t = np.linspace(-5, 5)
y = np.asarray([relu_function(i) for i in t ])
y2 = np.asarray([dev_relu_function(i) for i in t ])
title = 'ReLU Activation Function'
axis_displays2plot(t, y, t, y2, 'f(x)', 'f\'(x)', title)
```

## 6. Leaky ReLu:

\begin{equation*}
  f(x) = \left\{
    \begin{aligned}
      & 0.01x, x < 0 \\
      & x, x \geq  0
    \end{aligned}
  \right.
\end{equation*}

\begin{equation*}
  f'(x) = \left\{
    \begin{aligned}
      & 0.01, x < 0  \\
      & 1,  x \geq  0
    \end{aligned}
  \right.
\end{equation*}

### Key features:
- This function is a modified version of the ReLU function.
- Do not have the vanishing gradient problem.
- The leaky ReLU fucntion outputs a small negative value defined by $\alpha$z 
- Do not have any linear component with zero derivatives(slops). 
- The learning process there is faster than the default ReLU.

### Usage:
- The same usage of the ReLU function is also  valid for the leaky ReLU function.

```python
def leaky_relu_function(x):
    if x < 0:
        return 0.01*x
    else:
        return x
def dev_leaky_relu_functionn(x):
    if x < 0:
        return 0.01 
    else:
        return 1
    
t = np.linspace(-5, 5)
y = np.asarray([leaky_relu_function(i) for i in t ])
y2 = np.asarray([dev_leaky_relu_functionn(i) for i in t ])
title = 'Leaky ReLU Activation Function'
axis_displays2plot(t, y, t, y2, 'f(x)', 'f\'(x)', title)
```

## 7. Paramterised ReLU

\begin{equation*}
  f(x) = \left\{
    \begin{aligned}
      & ax, x < 0 \\
      & x, x \geq  0
    \end{aligned}
  \right.
\end{equation*}

\begin{equation*}
  f'(x) = \left\{
    \begin{aligned}
      & a, x < 0  \\
      & 1,  x \geq  0
    \end{aligned}
  \right.
\end{equation*}

where **"a"** is also a trainable parameter.

###  Key features:
- This is another variant of the ReLU function.
- Setting $\alpha$ as a parameter for each neuron in the network. How to optimate of the $\alpha$ learns from the network.

```python
def par_relu_function(a,x):
    if x < 0:
        return a*x
    else:
        return x
def dev_par_relu_functionn(a,x):
    if x < 0:
        return a
    else:
        return 1
    
t = np.linspace(-5, 5)
y = np.asarray([par_relu_function(3,i) for i in t ])
y2 = np.asarray([dev_par_relu_functionn(3,i) for i in t ])
title = 'Leaky ReLU Activation Function'
axis_displays2plot(t, y, t, y2, 'f(x)', 'f\'(x)', title)
```

## Exponential Linear Unit.

\begin{equation*}
  f(x) = \left\{
    \begin{aligned}
      & a(e^{x}-1), x < 0 \\
      & x, x \geq  0
    \end{aligned}
  \right.
\end{equation*}

\begin{equation*}
  f'(x) = \left\{
    \begin{aligned}
      & ae^{x}, x < 0 \\
      & 1, x \geq  0
    \end{aligned}
  \right.
\end{equation*}

```python
def elu_function(x, a):
    if x<0:
        return a*(np.exp(x)-1)
    else:
        return x
    
def dev_elu_function(x, a):
    if x<0:
        return a*np.exp(x)
    else:
        return 1

    
t = np.linspace(-5, 5)
y = np.asarray([elu_function(i,1) for i in t ])
y2 = np.asarray([dev_elu_function(i,1) for i in t ])
title = 'Exponential  Linear Unit Activation Function'
displays2plot(t, y, t, y2, 'f(x)', 'f\'(x)', title)
```

## Swish

$$f(x) = x*sigmoid(x)$$

$$f(x) = \frac{x}{1-e^{-x}}$$ 

### Key features:
- a non-linear function.
- Make of by multiplying the sigmoild function  by the inputs z.
- Have similaer among ReLU and Swisg.
- The curve is more smooth than the ReLU activation function. This smoothness is important when training the model. The function converges easily while training

### Usage:
- Only use in the hidden layer.
- We never use this function in the output layer of a neural network model.

### Drawbacks:
- The main drawback of the Swish function is that it is computationally expensive as an e^z term is included in the function. This can be avoided by using a special function called “Hard Swish” defined below.

```python
def sigmodess(x):
    return 1/(1-np.exp(-x)) 
def swish_function(x):
    return  x * sigmodess(x)

def dev_swish_function(x):
    return sigmodess(x) -swish_function(x)* np.exp(-x)*sigmodess(x)

t = np.linspace(-5, 5)
y = np.asarray([swish_function(i) for i in t ])
y2 = np.asarray([dev_swish_function(i) for i in t ])
title = 'Exponential  Linear Unit Activation Function'
displays2plot(t, y, t, y2, 'f(x)', 'f\'(x)', title)
```

## Softmax

$$ \sigma(z)_{j} = \frac {e^{z_{j}}}{\sum^{K}_{k = 1} e^{z_{k}}}$$  

for j = 1, ..., K

### Key features:
- This is also a non-linear activation function.
- This calculates the probability value  of a event over K different events. Sum of all value is 1, meaning that all events are mutually exclusive.

### Usage:
- Using for a multiclass classification problem.
- Never, usning in the hidden layer.
![imaeg](Image/2.png)

```python
def softmax_function(x):
    z = np.exp(x)
    z_ = z/z.sum()
    return z_

t = np.linspace(-5, 5)
y = softmax_function(t)
title = "softmax_function"
displays(t, y, title)
```
